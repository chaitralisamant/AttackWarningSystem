```python
parent = "/content/"
new_dir = "attacktest"

path = os.path.join(parent, new_dir)
os.mkdir(path)
```

```python
f = 1
for filename in os.scandir('attacks'):
  if filename.is_file():
    if f % 5 == 0:
      shutil.move(filename, 'attacktest')
      print(str(filename) + "to test set")
    f += 1
    print(filename)
```

```python
c=1
for filename in os.scandir('attacktest'):
    if filename.is_file():
      c += 1

print(c)
```
